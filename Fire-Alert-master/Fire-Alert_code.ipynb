```python
import pandas as pd

df = pd.read_csv("F-Alert1.csv") 

df.head(62)
```

```python
sqr=[]

for row in df.SecondCallReqd:
    if row in ['Yes']:
        sqr.append(1)
    elif row in ['No']:
        sqr.append(0)

df.insert(10, 'SCRBoolean', sqr)
```

```python
df.head(30)
```

```python
noc=[]

for row in df.CASUALITY:
    if row in ['No']:
        noc.append(0)
    else:
        noc.append(1)

df.insert(13, 'CasualityBoolean', noc)
```

```python
df.head(30)
```

```python
df['CallTime'] = pd.to_datetime(df.CallTime)
df['ReachTime'] = pd.to_datetime(df.ReachTime)
df['ReturnTime'] = pd.to_datetime(df.ReturnTime)
df.head()
```

```python
for row in df.CallTime:

    callth=df.CallTime.dt.hour
    calltm=df.CallTime.dt.minute
    
    reachth=df.ReachTime.dt.hour
    reachtm=df.ReachTime.dt.minute

    returnth=df.ReturnTime.dt.hour
    returntm=df.ReturnTime.dt.minute
    
    exth = returnth-reachth-(reachth-callth)
    extm = returntm-reachtm-(reachtm-calltm)
    
    extime = extm + (exth*60)
    
    df['ExtinguishTime']= extime

df.head(30)
```

```python
df['ExtinguishTime'].max()
```

```python
import numpy as np

df['ExtinguishTime'] = np.where(df['ExtinguishTime']<=0, 10, df.ExtinguishTime)
        
df.head(30)
```

```python
tmp_ext =df.ExtinguishTime - df.ExtinguishTime.min()
scaled_ext = tmp_ext / tmp_ext.max()
df["ScaledExt"] = scaled_ext
df.head(30)
```

```python
ebb=[]

for row in df.ExtinguishedBy:
    if row in ['Civilians']:
        ebb.append(0)
    else:
        ebb.append(1)


df.insert(16, 'ExtinguishedByBoolean', ebb)
```

```python
scrw = 0
cw = 0
ebw = 0

count = len(df.index)

scr_c = df['SCRBoolean'].sum()
c_c = df['CasualityBoolean'].sum()
eb_c = df['ExtinguishedByBoolean'].sum()

scrw = scr_c * math.log(count/scr_c)
cw = c_c * math.log(count/c_c)
ebw = eb_c * math.log(count/eb_c)
    
for row in df.CallTime:
    
    ss = (scrw * df.SCRBoolean) + (cw * df.CasualityBoolean) + (ebw * df.ExtinguishedByBoolean)
    
    df['SeverityScore']= ss

df.head(30)
```

```python
for row in df.CallTime:
    callth=df.CallTime.dt.hour
    calltm=df.CallTime.dt.minute
    
    calltime = calltm + (callth*60)
    
    df['CallTimeForClustering']= calltime

df.head(30)
```

```python
df['CallTimeForClustering'].max()
```

```python
df['CallTimeForClustering'] = pd.to_numeric(df.CallTimeForClustering)
df['ExtinguishTime'] = pd.to_numeric(df.ExtinguishTime)
df.head(30)
```

```python
tmp_clust =df.CallTimeForClustering - 0
scaled_clust = tmp_clust / 1440
df["ScaledClust"] = scaled_clust * 24
df.head(30)
```

```python
%matplotlib inline
import numpy as np
from pylab import *

def de_mean(x):
    xmean = mean(x)
    return [xi - xmean for xi in x]

def covariance(x, y):
    n = len(x)
    return dot(de_mean(x), de_mean(y)) / (n-1)

def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x,y) / stddevx / stddevy  #In real life you'd check for divide by zero here

scatter(df.ExtinguishedByBoolean,df.SeverityScore)

correlation (df.ExtinguishedByBoolean,df.SeverityScore)
```

```python
%matplotlib inline

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from numpy import random, float

data = df2.values[:,:]

model = KMeans(n_clusters=6)

# Note I'm scaling the data to normalize it! Important for good results.
model = model.fit(scale(data))

# We can look at the clusters each data point was assigned to
print(model.labels_)

# And we'll visualize it:
plt.figure(figsize=(8, 6))
plt.scatter(data[:,1], data[:,0], c=model.labels_.astype(float))
plt.show()
```

```python
print(model.labels_)

clust = model.labels_

df.insert(22,'Cluster',clust)

df.head(30)
```
